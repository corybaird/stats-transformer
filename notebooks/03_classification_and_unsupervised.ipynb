# Unsupervised Learning with Stats-Transformer
This notebook demonstrates PCA and clustering workflows.

```python
import pandas as pd
import numpy as np
from stats_transformer.models.unsupervised.unsupervised import PCAModel
from stats_transformer.visualization.eda.data_viz import DataVisualizer
```

## 1. Data Loading
Generate sample multivariate data.

```python
np.random.seed(42)
df = pd.DataFrame(np.random.randn(100, 5), columns=[f'feat_{i}' for i in range(5)])
df.head()
```

## 2. PCA Fitting
Fit a PCA model to reduce dimensionality.

```python
model = PCAModel(
    features=df.columns.tolist(),
    n_components=2
)
model.fit(df)
transformed_df = model.df_clean
print('Explained Variance Ratio:', model.get_model_metrics()['explained_variance_ratio'])
```

## 3. Visualization
Visualize the PCA results.

```python
viz = DataVisualizer(output_dir='data/notebook_exports/notebook_viz')
pca_paths = viz.create_scatter_plots(transformed_df, feature_list=['pca_1'], target='pca_2')
print('Visualizations complete')
```
