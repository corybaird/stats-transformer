# Time Series Analysis with Stats-Transformer
This notebook demonstrates how to fit time-series models like VAR.

```python
import pandas as pd
import numpy as np
from stats_transformer.models.timeseries.var import VARModel
from stats_transformer.visualization.models.model_viz import ModelVisualizer
```

## 1. Data Loading
Load sample time series data.

```python
from stats_transformer.data import load_sample_data

# Let's filter for one country to make it a pure time series
df = load_sample_data()
df = df[df['country'] == 'USA'].sort_values('date').dropna()
df.head()
```

## 2. Model Fitting
Fit a VAR model.

```python
model = VARModel(
    target_variables=['gdp', 'inflation'],
    maxlags=2,
    date_column='date'
)
metrics = model.fit(df)
print(model.get_summary())
```

Visualize the VAR results.
